```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.random.rand(20)
y = np.random.rand(20)

plt.scatter(x, y)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Homework 1 - Scatter Plot of Random Values')

plt.show()

plt.savefig('hw1_scatter.png')
```

```python
def find_max(k):
    return max(k)

numbers = [3, 7, 1, 9, 4]
print(find_max(numbers))
```

```python
def sort_list(i):
    return sorted(i)

numbers = [3, 5, 1, 9, 2]
print(sort_list(numbers))
```
